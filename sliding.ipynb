```python
from scipy.integrate import simps

def calc_power_in_frequency_band(eeg_data_group, sampling_rate):
    # Calculate fft result of voltage data
    fft_result = np.fft.fft(eeg_data_group)

    # Calculate the Power Spectral Density (PSD)
    psd = (1 / len(eeg_data_group)) * np.abs(fft_result) ** 2

    # Calculate the corresponding frequencies for the DFT result
    frequencies = np.fft.fftfreq(len(fft_result), 1 / sampling_rate)
    
    # Find the indices of frequencies within the specified frequency bands
    frequencies = frequencies[:len(frequencies) // 2]
    theta_freq_indices = np.where((frequencies >= 4) & (frequencies < 8))[0]
    alpha_freq_indices = np.where((frequencies >= 8) & (frequencies <= 13))[0]
    beta_freq_indices = np.where((frequencies >= 13) & (frequencies < 30))[0]
    gamma_freq_indices = np.where((frequencies >= 30) & (frequencies < 100))[0]
    
    # Find the peak frequency within the specified frequency bands
    theta_peak_frequency = frequencies[theta_freq_indices[np.argmax(psd[theta_freq_indices])]]
    alpha_peak_frequency = frequencies[alpha_freq_indices[np.argmax(psd[alpha_freq_indices])]]
    beta_peak_frequency = frequencies[beta_freq_indices[np.argmax(psd[beta_freq_indices])]]
    gamma_peak_frequency = frequencies[gamma_freq_indices[np.argmax(psd[gamma_freq_indices])]]

    # Find the minimum frequency within the specified frequency bands
    theta_min_frequency = frequencies[theta_freq_indices[np.argmin(psd[theta_freq_indices])]]
    alpha_min_frequency = frequencies[alpha_freq_indices[np.argmin(psd[alpha_freq_indices])]]
    beta_min_frequency = frequencies[beta_freq_indices[np.argmin(psd[beta_freq_indices])]]
    gamma_min_frequency = frequencies[gamma_freq_indices[np.argmin(psd[gamma_freq_indices])]]

    # Calculate the power within the specified frequency bands
    theta_power_in_freq_band = np.sum(psd[theta_freq_indices])
    alpha_power_in_freq_band = np.sum(psd[alpha_freq_indices])
    beta_power_in_freq_band = np.sum(psd[beta_freq_indices])
    gamma_power_in_freq_band = np.sum(psd[gamma_freq_indices])

    # Calculate the area under the FFT curve within the specified frequency bands using numerical integration
    theta_area_under_fft = simps(psd[theta_freq_indices], dx=1 / sampling_rate)
    alpha_area_under_fft = simps(psd[alpha_freq_indices], dx=1 / sampling_rate)
    beta_area_under_fft = simps(psd[beta_freq_indices], dx=1 / sampling_rate)
    gamma_area_under_fft = simps(psd[gamma_freq_indices], dx=1 / sampling_rate)

    area_under_fft = simps(psd[0:], dx=1 / sampling_rate)

    return (
        theta_peak_frequency, alpha_peak_frequency, beta_peak_frequency, gamma_peak_frequency,
        theta_min_frequency, alpha_min_frequency, beta_min_frequency, gamma_min_frequency,
        theta_power_in_freq_band, alpha_power_in_freq_band, beta_power_in_freq_band, gamma_power_in_freq_band,
        theta_area_under_fft, alpha_area_under_fft, beta_area_under_fft, gamma_area_under_fft,
        area_under_fft
    )
```
